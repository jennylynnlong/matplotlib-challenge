## Observations and Insights 

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np
from scipy.stats import linregress

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single dataset
study_data_complete = pd.merge(study_results, mouse_metadata, how="left", on="Mouse ID")

# Display the data table for preview
study_data_complete.head()
```

```python
# Checking the number of mice
MiceNo = study_data_complete.groupby('Mouse ID').nunique()
len(MiceNo)
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint.
duplicate_mouseID = study_data_complete.loc[study_data_complete.duplicated(subset=['Mouse ID', 'Timepoint']), 'Mouse ID'].unique()
#duplicate_mouseID

count_of_mouseID = study_data_complete.groupby('Mouse ID')['Timepoint'].count()

# create dataframe
count_of_mouseDF = pd.DataFrame(
    {
        "Counts of Mouse ID": count_of_mouseID
    }
)

# sort data frame in descending values
count_of_mouseDF = count_of_mouseDF.sort_values(by="Counts of Mouse ID", ascending=False)

# remove index header
count_of_mouseDF.index.name = None
count_of_mouseDF.head()
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
duplicateMouse = study_data_complete.loc[study_data_complete['Mouse ID'] == 'g989',:]

# set the index to the mouse ID
duplicateMouse = duplicateMouse.set_index('Mouse ID')

# check the mouse data for ID
duplicateMouse
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
# need to use drop.duplicate with both 'Mouse ID' and 'Timepoint' and keep=first
cleanDF = study_data_complete.drop_duplicates(subset=['Mouse ID', 'Timepoint'], keep='first')
cleanDF.head()
```

```python
# Checking the number of mice in the clean DataFrame.
len(cleanDF['Mouse ID'].unique())
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
mean = cleanDF.groupby('Drug Regimen').mean()['Tumor Volume (mm3)']
median = cleanDF.groupby('Drug Regimen').median()['Tumor Volume (mm3)']
variance = cleanDF.groupby('Drug Regimen').var()['Tumor Volume (mm3)']
stdev = cleanDF.groupby('Drug Regimen').std()['Tumor Volume (mm3)']
SEM = cleanDF.groupby('Drug Regimen').sem()['Tumor Volume (mm3)']

# Assemble the resulting series into a single summary dataframe.

StatsTable = pd.DataFrame(
    {
        "Mean Tumor Volume": mean,
        "Median Tumor Volume": median,
        "Tumor Volume Variance": variance,
        "Tumor Volume Stdev": stdev,
        "Tumor Volume SEM": SEM
    }
)

# Preview DataFrame
StatsTable
```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Using the aggregation method, produce the same summary statistics in a single line
aggreTable = cleanDF.groupby("Drug Regimen").agg(
    Mean_Tumor_Volume=('Tumor Volume (mm3)',np.mean),
    Median_Tumor_Volume=('Tumor Volume (mm3)',np.median),
    Variance_Tumor_Volume=('Tumor Volume (mm3)',np.var),
    Stdev_Tumor_Volume=('Tumor Volume (mm3)',np.std),
    SEM_Tumor_Volume=('Tumor Volume (mm3)',st.sem)
)
aggreTable
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin

# set drug regimen as index and drop associated regimens while only keeping Capomulin, Ramicane, Infubinol, and Ceftamin
drugDF = cleanDF.set_index('Drug Regimen')
drugDF = drugDF.loc[['Capomulin', 'Ramicane', 'Infubinol', 'Ceftamin'],:]

# isolated view of just capomulin for later use
capDF = drugDF.loc['Capomulin',:]

# Reset index so drug regimen column persists after inner merge
drugDF = drugDF.reset_index()

# get mouse count per drug
newMouseCount = drugDF["Drug Regimen"].value_counts()
#print(newMouseCount)

# Start by getting the last (greatest) timepoint for each mouse
lastTumorVol = drugDF.groupby('Mouse ID')['Timepoint'].max()
lastTumorVol.head()

#drugDF.head()
```

```python
# reset lastTumorVol index
lastTumorVol = lastTumorVol.reset_index()

# Merge this group df with the original dataframe to get the tumor volume at the last timepoint
mergedDF = lastTumorVol.merge(cleanDF, on=['Mouse ID', 'Timepoint'], how='left')
mergedDF = mergedDF.set_index('Mouse ID')

# show all rows of data
mergedDF = mergedDF.sort_values(by='Timepoint')
mergedDF = mergedDF[['Drug Regimen', 'Timepoint', 'Tumor Volume (mm3)', 'Metastatic Sites', 'Sex', 'Age_months', 'Weight (g)']]
pd.set_option('display.max_rows', None)
mergedDF
```

```python
# Put treatments into a list for for loop (and later for plot labels)
# print(mergedDF['Drug Regimen'].unique())
txList = ['Ceftamin', 'Infubinol', 'Ramicane', 'Capomulin']

# Create empty list to fill with tumor vol data (for plotting)
tumorList = []

# Calculate the IQR and quantitatively determine if there are any potential outliers. 
for t in txList:
    # Locate the rows which contain mice on each drug and get the tumor volumes
    finalTumorCount = mergedDF.loc[mergedDF['Drug Regimen'] == t, 'Tumor Volume (mm3)']
    
    # append to tumorList
    tumorList.append(finalTumorCount)
    
    # determine outliers
    quartiles = finalTumorCount.quantile([0.25, 0.5, 0.75])
    lowerQ = quartiles[0.25]
    upperQ = quartiles[0.75]
    IQR = upperQ - lowerQ
    lowerbound = lowerQ - (1.5 * IQR)
    upperbound = upperQ + (1.5 * IQR)
    outliers = finalTumorCount.loc[(finalTumorCount < lowerbound) | (finalTumorCount > upperbound)]
    print(f'The potential outliers for {t} are: {outliers}')
```

```python
# Determine outliers overall using upper and lower bounds

Tumors = mergedDF['Tumor Volume (mm3)']
overall_quartiles = Tumors.quantile([0.25, 0.5, 0.75])
overall_lowerQ = overall_quartiles[0.25]
overall_upperQ = overall_quartiles[0.75]
overall_IQR = overall_upperQ - overall_lowerQ
overall_lowerbound = overall_lowerQ - (1.5 * overall_IQR)
overall_upperbound = overall_upperQ + (1.5 * overall_IQR)

print(f'The lower quartile of Tumor Volume is: {overall_lowerQ}')
print(f'The lower quartile of Tumor Volume is: {overall_upperQ}')
print(f'The interquartile range of Tumor Volume is: {overall_IQR}')
print(f'The median of Tumor Volume is: {mergedDF["Tumor Volume (mm3)"].median()}')
print(f'Values below {overall_lowerbound} could be outliers.')
print(f'Values above {overall_upperbound} could be outliers.')
```

```python
# set color for the plots, and markers
red_out = dict(markerfacecolor = 'red', markersize=7)

# box plot tumor volumes for each Drug Regimen
plt.boxplot(tumorList, labels=txList, flierprops=red_out)
plt.xlabel('Drug Regimen')
plt.ylabel('Tumor Volume')
plt.title('Tumor Volume For Drug Regimen')
plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and linear regression model 
# for mouse weight and average tumor volume for the Capomulin regimen
#establish x and y values and find St. Pearson Correlation Coefficient for Mouse Weight and Tumor Volume Avg
x_weight = cap_avg['Weight (g)']
y_tumorVol = cap_avg['Tumor Volume (mm3)']
corr = st.pearsonr(x_weight,y_tumorVol)

#print St. Pearson Correlation Coefficient
print(f'The correlation between both factors is {round(corr[0],2)}')

# establish linear regression values
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_weight, y_tumorVol)
regress_values = x_weight * slope + intercept


# linear regression line 
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))

# scatter plot of the data
plt.scatter(x_weight, y_tumorVol)
plt.plot(x_weight, regress_values, "r-")
plt.annotate(line_eq, (18,36), fontsize=15, color='red')
plt.xlabel('Mouse Weight (g)')
plt.ylabel('Tumor Volume Average')
plt.title('Tumor Volume Average vs. Mouse Weight Linear Regression')
plt.show()
```

Three observations:

    1) It looks as though there is a correlation between how big a 
    tumor is and how big the mouse is. The larger the mouse, the larger
    the tumor. The r-squared value is 0.84 which proves there is a 
    strong correlation between tumor volume and mouse weight.
    
    2) Out of the drug regimens pulled out of the data (Ceftamin, 
    Infubinol, Ramicane, and Capomulin), Capomulin and Ramicane appear
    to be the most effective at decreasing or maintaining tumor volume
    in mice.
    
    3) Capomulin and Ramicane are the two lowest drug regimens amongst
    all drugs used in this study for standard deviation and variance
    which shows that these drugs are consistent in their effect on 
    tumors. 
    
    4) The study was evenly tested amongst all of the drug regimens as
    well as between male and female mice. It would be interesting to see 
    the correlation between age and tumor volume to see if younger mice 
    see better results from all or certain drug regimens than older
    mice.
